```python
def solo_fecha(registro):
  return registro.date()
```

```python
cuv_cero = pd.read_excel('/content/drive/MyDrive/4_Python/1_Sandbox/Taranto/CUV Articulos por division 202201.xlsx')
cuv_cero.rename(str.lower, axis="columns", inplace=True)

print(f"El primer dataset tiene {cuv_cero.shape[0]} filas y {cuv_cero.shape[1]} columnas \nLas cuales son: {cuv_cero.columns.tolist()}")
print(f"Hay {cuv_cero['division'].nunique()} divisiones, las cuales son: {cuv_cero['division'].unique().tolist()}")
```

```python
cuv_uno = pd.read_excel("/content/drive/MyDrive/4_Python/1_Sandbox/Taranto/CUV Articulos por division 202202.xlsx")
cuv_uno.rename(str.lower, axis="columns", inplace=True)

print(f"El segundo dataset tiene {cuv_cero.shape[0]} filas y {cuv_cero.shape[1]} columnas \nLas cuales son: {cuv_uno.columns.tolist()}")
print(f"Hay {cuv_uno['division'].nunique()} divisiones, las cuales son: {cuv_uno['division'].unique().tolist()}")
```

```python
cuv_viejos = cuv_cero[cuv_cero["division"].isin(["estampado", "fijacion", "forja", "sellado"])]
cuv_viejos.columns = cuv_viejos.columns.str.replace("cuv", "").str.strip("_")+ "_a"
# cuv_viejos.columns = cuv_viejos.columns.str.replace("cuv", "").str.replace("_","")+ "_a"

print(cuv_viejos.tail(3))
print(f"\n{cuv_viejos.shape}")
```

```python
cuv_nuevos = cuv_uno[cuv_uno["division"].isin(["estampado", "fijacion", "forja", "sellado"])]
cuv_nuevos.columns = cuv_nuevos.columns.str.replace("cuv", "").str.strip("_") + "_b"
# cuv_nuevos.columns = cuv_nuevos.columns.str.replace("cuv", "").str.replace("_","") + "_b"

print(cuv_nuevos.tail(3))
print(f"\n{cuv_nuevos.shape}")
```

```python
cuv_viejos_forja = pd.read_excel("/content/drive/MyDrive/4_Python/1_Sandbox/Taranto/CUV Articulos por division 202201.xlsx", "FORJA").rename(str.lower, axis=1)
cuv_viejos_forja.columns = cuv_viejos_forja.columns.str.replace("cuv", "").str.strip("_") + "_a"
# print(cuv_viejos_forja["vigencia_cuv"].dtype)

cuv_viejos_forja["vigencia_a"] = pd.to_datetime(cuv_viejos_forja["vigencia_a"], errors="coerce")
print(cuv_viejos_forja.loc[cuv_viejos_forja["vigencia_a"].isnull()])
print(f"\n{cuv_viejos_forja.head(3)}")
print(f"\n{cuv_viejos_forja.shape}")
```

```python
cuv_nuevos_forja = pd.read_excel("/content/drive/MyDrive/4_Python/1_Sandbox/Taranto/CUV Articulos por division 202202.xlsx", "FORJA").rename(str.lower, axis=1)
cuv_nuevos_forja.columns = cuv_nuevos_forja.columns.str.replace("cuv", "").str.strip("_") + "_b"

cuv_nuevos_forja["vigencia_b"] = pd.to_datetime(cuv_nuevos_forja["vigencia_b"], errors="coerce")
print(cuv_nuevos_forja.loc[cuv_nuevos_forja["vigencia_b"].isnull()])
print(f"\n{cuv_nuevos_forja.head(3)}")
print(f"\n{cuv_nuevos_forja.shape}")
```

```python
prueba_0 = pd.concat([cuv_viejos, cuv_viejos_forja], axis = 0)

for i in ["vigencia_a"]:
  prueba_0[i] = prueba_0[i].map(solo_fecha)

prueba_0
```

```python
prueba_1 = pd.concat([cuv_nuevos, cuv_nuevos_forja], axis = 0)

for i in ["vigencia_b"]:
  prueba_1[i] = prueba_1[i].map(solo_fecha)

prueba_1
```

```python
# cuenta piezas duplicadas en ambos dfs

count_dups_a = ((prueba_0["codigo_a"].duplicated()) == True).sum()
count_dups_b = ((prueba_1["codigo_b"].duplicated()) == True).sum()

print(f"La cantidad de piezas duplicadas en el df anterior (viejo) es de {count_dups_a}.")
if count_dups_a == count_dups_b:
  print(f"En el nuevo es también de {count_dups_b}.")
else:
  print(f"En el nuevo es de {count_dups_b}.")

# ALTERNATIVA -> "duplicated" no te dice cantidad de duplicados, solo devuelve un array de booleanos. El "any" te dice si hay o no valores duplicados
# boolean = df.duplicated().any() # look in the df as a whole, row wise
# boolean = df.duplicated(subset=['col1','col2']).any() # for specific columns. There's also a keep parameter
```

```python
cuv_viejos_mix = prueba_0.drop_duplicates(subset="codigo_a", keep="last", ignore_index=True)
cuv_nuevos_mix = prueba_1.drop_duplicates(subset="codigo_b", keep="last", ignore_index=True)

print(cuv_viejos_mix.shape)
print(cuv_nuevos_mix.shape)
```

```python
# Info:
# https://stackoverflow.com/questions/39985861/pandas-merge-on-columns-with-different-names-and-avoid-duplicates
# https://pandas.pydata.org/docs/user_guide/merging.html#joining-key-columns-on-an-index

comparar_0 = pd.merge(cuv_viejos_mix.set_index("codigo_a"), cuv_nuevos_mix, how="outer", left_index=True, right_on=["codigo_b"]).drop(["division_a"], axis=1)
print(f"{comparar_0.shape}\n")
comparar_0.head(3)
```

```python
# Alternative:

comparar_0_bis = pd.merge(cuv_viejos_mix, cuv_nuevos_mix, how="outer", left_on=["codigo_a"], right_on=["codigo_b"]).drop(["codigo_a", "division_a"], axis=1)
print(f"{comparar_0_bis.shape}\n")
comparar_0_bis.head(3)
```

```python
orden_alfa = comparar_0.columns.tolist()
print(f"Hay {len(orden_alfa)} elementos/columnas hasta ahora en la lista\n")
print(f"Los cuales son: {orden_alfa}\n")
# Orden deseado alternativa 1:
orden_nuevo_alfa = [3, 7, 1, 0, 2, 5, 4, 6]
# Orden deseado alternativa 2:
orden_nuevo_alfa_bis = ["codigo_b", "division_b", "vigencia_a", "usd_a", "segundos_a", "vigencia_b", "usd_b", "segundos_b"]

orden_final_alfa = [orden_alfa[i] for i in orden_nuevo_alfa]
print(f"Pero quiero este orden: {orden_final_alfa}")

if orden_final_alfa == orden_nuevo_alfa_bis:
  print("\nCheck: ambas listas para el nuevo orden son iguales")
else:
  print("\nCheck: las listas para el nuevo orden no son iguales")
```

```python
comparar_1 = comparar_0.reindex(orden_final_alfa, axis=1)

# https://docs.python.org/3/library/decimal.html
# https://stackoverflow.com/questions/1995615/how-can-i-format-a-decimal-to-always-show-2-decimal-places
# https://datascienceparichay.com/article/remove-time-from-date-pandas/
# https://python-forum.io/thread-32328.html

comparar_1["dias_transcurridos"] = comparar_1["vigencia_b"] - comparar_1["vigencia_a"]
# ((comparar_1["vigencia_b"] - comparar_1["vigencia_a"]) / 30)

print(comparar_1[["vigencia_a", "vigencia_b"]].dtypes)

comparar_1["var_%_usd"] = ((comparar_1["usd_b"] / comparar_1["usd_a"] - 1) * 100).round(1)
comparar_1[["segundos_a", "segundos_b"]] = comparar_1[["segundos_a", "segundos_b"]].round(3) # redondeado a 3 decimales para ver en milisegundos
comparar_1["var_%_seg"] = ((comparar_1["segundos_b"] / comparar_1["segundos_a"] - 1) * 100).round(1)

print(f"\n{comparar_1.shape}")
comparar_1.head(3)
```

```python
oem_parts = pd.read_excel("/content/drive/MyDrive/4_Python/1_Sandbox/Taranto/OEM_Parts_PAG_2023.xlsx")
oem_parts.rename(str.lower, axis="columns", inplace=True)
oem_parts.drop(columns=["división"], inplace=True)
oem_parts.sample(3)
```

```python
comparar_2 = pd.merge(comparar_1.rename(columns={"codigo_b":"pieza", "division_b":"division"}).sort_values(by=["var_%_usd"], ascending=False),
                                     oem_parts,
                                     how="inner",
                                     on="pieza"
                                     )

# Las columnas quedan así: ['pieza', 'division', 'vigencia_a', 'usd_a', 'segundos_a', 'vigencia_b', 'usd_b', 'segundos_b', 'dias_transcurridos', 'var_%_usd', 'var_%_seg', 'cliente']
orden_beta = comparar_2.columns.to_list()
# print(f"{pd.Series(orden_beta)}\n")

# Pero necesito que el "cliente" vaya luego de la "división", es decir: [0, 1, 11, 2, 3 , 4, 5, 6, 7, 8, 9, 10]
orden_nuevo_beta = [0, 1, 11, 2, 3 , 4, 5, 6, 7, 8, 9, 10]
orden_final_beta = [orden_beta[i] for i in orden_nuevo_beta]

comparar_3 = comparar_2.reindex(orden_final_beta, axis="columns")
comparar_3.head(3)
```

```python
# https://stackoverflow.com/questions/67894360/writing-pandas-dataframe-to-excel-how-to-auto-adjust-column-widths
# https://stackoverflow.com/questions/55280131/no-module-named-xlsxwriter-error-while-writing-pandas-df-to-excel
# https://stackoverflow.com/questions/69263078/pandas-dataframe-to-excel-cell-alignment
# https://stackoverflow.com/questions/17232013/how-to-set-the-pandas-dataframe-data-left-right-alignment
# https://datascientyst.com/change-display-options-pandas-styler-set_properties/
# https://pandas.pydata.org/pandas-docs/stable/user_guide/style.html

writer = pd.ExcelWriter("/content/drive/MyDrive/4_Python/1_Sandbox/Taranto/CUV_202202_vs_202201.xlsx", engine="xlsxwriter")
comparar_3.style.set_properties(**{"text-align":"center"}).to_excel(writer, sheet_name="info", index=False, freeze_panes=(2, 3))
workbook = writer.book
worksheet = writer.sheets["info"]
for i, col in enumerate(comparar_3.columns):
  width = max(comparar_3[col].apply(lambda x: len(str(x))).max(), len(col)) + 1
  worksheet.set_column(i, i, width)
writer.save()
```

```python
'https://stackoverflow.com/questions/72642329/is-there-a-way-to-export-a-dataframe-of-multiple-column-index-without-the-row-in'

from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows

# opening an excel workbook and worksheet
wb = Workbook()
ws = wb.active

lab2 = comparar_3

# writing dataframe to excel
for r in dataframe_to_rows(lab2, index=False, header=True):
    ws.append(r)

# merging header cells
for merge in range(int(comparar_3.shape[1]/3)):
    ws.merge_cells(start_row=1, end_row=1, start_column=merge*3+1, end_column=merge*3+3)

print(lab2.info())

# saving to excel
wb.save("/content/drive/MyDrive/4_Python/1_Sandbox/Taranto/testino.xlsx")
```
