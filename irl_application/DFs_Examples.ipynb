```python
# The enumerate () method adds a counter to an iterable and returns it in the form of an enumerating object.
# This enumerated object can then be used directly for loops or converted into a list of tuples using the list() function.
# Parameters:
# Iterable: any object that supports iteration
# Start: the index value from which the counter is to be started, by default it is 0

y = ["casa", "sol"]
print(type(enumerate(y)))
print(list(enumerate(y)))
print(list(enumerate(y, 100)), "\n")

for ele in y:
  print(ele)

print("\n")

for ele in enumerate(y):
  print(ele)

for count, ele in enumerate(y):
  print("\n", count, ele) # separately
  print(count)
  print(ele)
```

```python
def create_matrix_alfa(rows, cols, value=0):
    return [[value for _ in range(cols)] for _ in range(rows)]

def is_matrix_alfa(matrix_alfa):
    if not matrix_alfa or not isinstance(matrix_alfa, list) or not all(isinstance(row, list) for row in matrix_alfa):
        return False
    row_length = len(matrix_alfa[0])
    return all(len(row) == row_length for row in matrix_alfa)

matrix_alfa = create_matrix_alfa(3, 4)
print("matrix_alfa:", matrix_alfa)
print("Is this a matrix_alfa?", is_matrix_alfa(matrix_alfa))
```
