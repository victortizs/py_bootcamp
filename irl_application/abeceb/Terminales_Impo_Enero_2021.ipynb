```
df1 = pd.read_excel('/content/drive/MyDrive/Popis Fam/Terminales1.xlsx')
df1
```

```
df2 = pd.read_excel('/content/drive/MyDrive/Popis Fam/destinacion_impo.xls')
df2 = df2.rename(columns = {'Codigo' : 'DESTINAC'})
df2
```

```
terminales = pd.merge(df, df1, how='inner', on='CUIT')
terminales
```

```
terminales_destinacion = pd.merge(terminales, df2, how='inner', on='DESTINAC')
terminales_destinacion
```

```
a = []
s = []
o = []

for i in terminales_destinacion['NCM'].astype(str):
  a.append(i)

for i in a:
  s.append(i[0:4])
  o.append(i[0:8])
```

```
terminales_destinacion['NCM4'] = s
terminales_destinacion['NCM8'] = o
terminales_destinacion
```

```
suma_terminal_destinacion = terminales_destinacion.groupby(['AÑO','MES','Terminal','ORIGEN','NCM4'], as_index=False).agg({'CIF_USS':'sum',
                                                                                                                          'CANTIDAD':'sum',
                                                                                                                          'KGS_NETOS':'sum'})
suma_terminal_destinacion
```
