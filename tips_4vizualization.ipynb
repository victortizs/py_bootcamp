## Other tips/tricks

```python
# Control decimal places to column. I won't run this cell either

pd.set_option('precision', 2) # 2 is the number of decimals I choose to see in the dataset
```

```python
# The sequence of items (lists) are also truncated if they have many characters:

df1["New col"] = [['a']*100] * df1.shape[0] # df1.shape[0] it's to match the number of rows the dataset has
df1[["Fecha", "New col"]].head()
```

```python
pd.reset_option('max_colwidth')
df1[['Fecha', 'New col']]

# But, if u haven't touched the "max_colwidth" option, this is what happens
```

```python
# So for display all the items (list) that a single register could contain:

pd.set_option("display.max_colwidth", None)
pd.set_option("max_seq_item", None)
df1[['Fecha', 'New col']]
```
