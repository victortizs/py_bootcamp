```python
df = pd.read_csv('/content/drive/MyDrive/CAEM/Muestra/Febrero/exponm21.csv', sep=";", encoding='latin-1') # leé este csv separado por ";" y codificado (info pasada a código) según esta región específica
df.head() # me muestra las primeras 5 filas del dataset por default, pero depende del n° que ponga dentro de los paréntesis
```

```python
tamaño = df.shape # me devuelve una tupla (tupla = conjunto de datos inmutables y ordenados) con el tamaño del dataset (filas;columnas)
print(f"El dataset tiene {tamaño[0]} filas y {tamaño[1]} columnas.") # 0 refiere al primer elemento de la tupla (filas) y 1 al segundo (columnas). Python cuenta desde cero
```

```python
df.info() # me dice si existen valores nulos y que tipo de datos son los valores de cada columna (en qué formato están)
```

```python
sin_confis = df.drop(df[df["Pnet(kg)"].str.contains(r's')].index) # elimina las filas que en la columna Pnet(kg) SÍ contegan 's'
sin_confis.sort_values(by=['Año','Mes','NCM'], ascending=True, inplace=True) # ordena los datos de menor a mayor y sobreescribe luego la tabla sin_confis (guarda los cambios)
sin_confis
```

```python
def reemplaza(numero): # armo una función (llamada reemplaza) que cambie en el argumento que elija (llamado numero) las "," por ".", dado que Python habla inglés
  return numero.replace(",",".")

for i in ['Pnet(kg)','FOB(u$s)']: # por cado elemento en estas columnas, aplica la función creada anteriormente
  sin_confis[i] = sin_confis[i].map(reemplaza)

sin_confis
```

```python
sin_confis[['Pnet(kg)','FOB(u$s)']] = sin_confis[['Pnet(kg)','FOB(u$s)']].apply(pd.to_numeric) # convertí los datos de estas columnas (antes objects) en número (float o integer según corresponda) por default
sin_confis
```

```python
writer = pd.ExcelWriter("/content/drive/MyDrive/CAEM/Sin confidenciales ejemplo.xlsx") # creo un Excel para guardar los datos obtenidos y elijo la ruta donde quiero tenerlo
sin_confis.to_excel(writer, sheet_name='Partidas', index=False) # elijo la tabla sin_confis para guardarla en dicho libro, pongo nombre a la hoja y no quiero ver el índice que acompaña a las filas de la tabla
writer.save() # lo guardo
```

```python
confis_paso_1 = df.drop(df[~df["Pnet(kg)"].str.contains(r's')].index) # elimina las filas del csv de INDEC que en la columna Pnet(kg) NO [~] contengan 's'
confis_paso_1 = confis_paso_1.drop(columns=['Pnet(kg)', 'FOB(u$s)','Pdes']) # borra estas columnas
confis_paso_1 = confis_paso_1.drop_duplicates(subset=['Año','Mes','NCM'], keep="first") # de los duplicados, quédate con el primero (es indistinto para el caso quedarse con el primero o el segundo)

# Ahora lo mismo pero con la tabla que contenía las partidas no confidenciales, la del comienzo
confis_paso_2 = sin_confis.drop(['Pnet(kg)', 'FOB(u$s)','Pdes'], axis=1) # drop([labels], axis=1) is equivalent to drop(columns=[labels])
confis_paso_2 = confis_paso_2.drop_duplicates(subset=['Año','Mes','NCM'], keep="first")
```

```python
confis = pd.concat([confis_paso_1, confis_paso_2], axis=0) # axis=0 agrega los datos de la 2da. tabla como filas adicionales (debajo de la primera); axis=1 agrega los datos como columnas adicionales (a la derecha de la primera)
confis = confis.drop_duplicates(keep=False) # si los datos no se presentan una única vez, elimínalos (no te quedes ni con la primera ni la segunda aparición)
confis.sort_values(by=['Año','Mes','NCM'], ascending=True, inplace=True) # ordena el dataset de menor a mayor según las columnas elegidas y sobreescribe la tabla confis [inplace=True]
confis
```

```python
writer1 = pd.ExcelWriter("/content/drive/MyDrive/CAEM/Muestra/Febrero/Confidenciales.xlsx") # creo un Excel para esta tabla, y la ruta donde se guardará
confis.to_excel(writer1, sheet_name="Partidas", index=False) # igual que antes
writer1.save() # igual que antes
'FIN'
```
